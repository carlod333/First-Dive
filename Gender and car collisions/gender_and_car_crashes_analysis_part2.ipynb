```python
# Run AutoML for 30 mins
aml = H2OAutoML(seed=333, max_runtime_secs=60*30, nfolds = 5, stopping_metric = "AUC")
aml.train(x = x, y = y, training_frame = train)
```

```python
# printing entire leaderboard of 121 models here - just to show how many models autoML considered!
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
# Evaluate performance of leader on test set
perf = aml.leader.model_performance(test)
perf
```

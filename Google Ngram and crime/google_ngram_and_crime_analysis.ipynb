```python
!pip install h2o
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
df = pd.read_csv('/content/drive/MyDrive/first dive/ngram_crime_data.csv', index_col=0)
```

```python
df = h2o.H2OFrame(df)
```

```python
train = df[df["year"] < 2008]
test = df[ df["year"] >= 2008]
```

```python
x = ["shit", "fuck", "cunt", "cocksucker", "motherfucker", "tits"]
y = "violent"
```

```python
# Run AutoML
aml = H2OAutoML(seed=333, max_models=15, max_runtime_secs=120, include_algos=["XGBoost", "GBM", "GLM", "DRF", "StackedEnsemble"])
aml.train(x = x, y = y, training_frame = train)
```

```python
# Print Leaderboard (ranked by validation set metrics)
aml.leaderboard
```

```python
aml.leader
```

```python
# Evaluate performance on a test set
perf = aml.leader.model_performance(test)
```
